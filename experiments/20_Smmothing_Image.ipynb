```python
import cv2
import matplotlib.pyplot as plt

# Load image
img = cv2.imread(r"C:\Users\Abhishek\OneDrive\Pictures\sunset.jpg")
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# Apply smoothing (averaging blur)
smooth = cv2.blur(img, (5,5))

# Show both images
plt.figure(figsize=(8,4))

plt.subplot(1,2,1)
plt.imshow(img)
plt.title("Original Image")

plt.subplot(1,2,2)
plt.imshow(smooth)
plt.title("Smoothed Image")

plt.show()
```
